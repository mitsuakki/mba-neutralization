## Introduction to operations and equivalence

As we said earlier an **MBA** expression is one that combine arithmetic operations such as (`ADD`,  `SUB`, `MUL`, etc..) with bitwise operations (`XOR`, `AND`, `NOT`, `OR`, etc..)<br> All this operations have a corresponding symbol to decribe them:


| **Operator** | **Symbol** | **Description**      |
|--------------|------------|----------------------|
| ADD          | +          | Addition             |
| SUB          | -          | Subtraction          |
| MUL          | *          | Multiplication       |
| XOR          | ^          | Exclusive OR (XOR)   |
| AND          | &          | Logical AND          |
| NOT          | ~          | Logical NOT          |
| OR           | |          | Logical OR           |

The concept of MBA allows to transform a simple expression into one that is more or less difficult to understand.


For exemple, this simple expression `x+y` is equivalent to this one `2*(-1)-(~x)-(~y)`, here's how:
1. Start with the given expression:
   $$
   \
   2 \cdot (-1) - (\sim x) - (\sim y)
   \
   $$

2. Evaluate `2 * (-1)`:
   $$
   \
   2 \cdot (-1) = -2
   \
   $$
   The expression becomes:
   $$
   \
   -2 - (\sim x) - (\sim y)
   \
   $$

3. Substitute the binary representation of `~x` and `~y`:  
   In binary logic, `~x = -x-1` and `~y = -y-1`.

   Substitute these into the expression:
   $$
   \
   -2 - (-x - 1) - (-y - 1)
   \
   $$

4. Simplify the terms:
   $$
   \
   -2 + x + 1 + y + 1
   \
   $$

   Combine constants:
   $$
   \
   x + y + 0
   \
   $$

5. Final result:
   $$
   \
   x + y
   \
   $$

Notice that this expression is also equivalent to (2*(x | y)) - (x ^ y) and many more that we probably doesn't know at the moment !! <br>

For each expressions with two variables `x` and `y`, it exists 28 basic combinations. <br>
With parentheses and several operators combined, the number of MBAs becomes potentially infinitely large.
